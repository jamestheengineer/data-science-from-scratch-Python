```
# Chapter 3 - Visualizing data. Two primary uses for visualizations: 1. To explore data
# 2. To communicate data

# Matplotlib is standard
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14959.3]

# Create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# Add a title
plt.title("nominal GDP")

# Add a label to the y-axis
plt.ylabel("Billions of $")
plt.show()
```

```
# Bar charts are good when showing discrete sets of items
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# Plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(movies)), num_oscars)
plt.title("My Favorite Movies")
plt.ylabel("# of Academy Awards")

# Label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)

plt.show()
```

```
# Bar charts are also good for plotting histograms of bucketed numeric values to see how things are distributed
from collections import Counter

grades = [83, 95, 91, 87, 70, 0 , 85, 82, 100, 67, 73, 77, 0]

# Bucket grades by decile, but put 100 in with the 90s
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades) # // is floor division

plt.bar([x + 5 for x in histogram.keys()],    # Shift bars right by 5
        histogram.values(),                   # Give each bar its correct height
        10,                                   # Give each bar a width of 10                      
        edgecolor=(0,0,0))                    # Black edges for each bar

plt.axis([-5, 105, 0, 5])   # x-axis from -5 to 105,
                            # y-axis from 0 to 5

plt.xticks([10 * i for i in range(11)])   # x-axis labels at 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

```
# Don't mislead people with your axes:

mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.-13e3 off in the corner (bad matplotlib!)
plt.ticklabel_format(useOffset=False)

# misleading y-axis only shows the part above 500
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Look at the 'Huge' Increase")
plt.show()
```

```
# Here's something a little more sensible
mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.-13e3 off in the corner (bad matplotlib!)
plt.ticklabel_format(useOffset=False)
plt.axis([2016.5, 2018.5, 0, 550])
plt.title("Not so Huge Anymore")
plt.show()
```
